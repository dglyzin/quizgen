```python
#generate samples
def selection_gen(accountfname, ucount, resfilename, excludefname=None):
    from collections import OrderedDict
    import numpy as np
    import json
    
    if excludefname:
        with open(excludefname, "r") as studentfile:
            oldstudlistdict = json.load(studentfile)

    
    tcount = len(globaltasks)

    accfile = open(accountfname, "r")
    acclist = [[line.rstrip().split(" ")[0], line.rstrip().split(" ")[6], line.rstrip().split(" ")[3], line.rstrip().split(" ")[4], line.rstrip().split(" ")[5]  ] for line in accfile]
    accfile.close()

    studlistdict = OrderedDict([])
    for student in acclist:
        studuid = student[0]
        if excludefname:
            oldstudinfo = oldstudlistdict[studuid]
        studgroup = student[1]
        studfio = " ".join(student[2:5])
        #generate problem indices
        choice = list(np.random.choice(range(1,tcount+1), ucount, replace=False))
        studproblems=[]
        for selectedidx in choice:
            taskgroup = globaltasks[selectedidx-1]
            #generate variant index
            if excludefname:
                #find variant in previous attempt
                for oldprob in oldstudinfo["problems"]:
                    if oldprob["problemidx"] == selectedidx:
                        oldvariantidx =  oldprob["variantidx"]
                
                variantidx = (oldvariantidx % len(taskgroup)) + 1 
            else:
                variantidx = (np.random.randint(1,len(taskgroup)+1))
            taskvariant = taskgroup[variantidx-1]
            #generate parameter choice
            taskparams = taskvariant['params']
            chosenparams=OrderedDict([])
            for param in taskparams: 
                selparamidx = np.random.randint(0,len(taskparams[param]))
                chosenparams.update([(param,taskparams[param][selparamidx])])    
            problemdict = {"problemidx":int(selectedidx), "variantidx":variantidx, "params":chosenparams}
            studproblems.append(problemdict)

        studinfo = {"group":studgroup, "fio":studfio, "problems":studproblems}    
        studlistdict.update([(studuid, studinfo)])

    #print(studlistdict)    
    with open(resfilename, "w") as selfile:
        json.dump(studlistdict, selfile,  indent=4)
#==================#
selection_gen("accounting/userinputlist.txt", 10, "selection.txt")
#selection_gen("accounting/userinputlist.txt", 10, "selection.txt", "selection_old.txt")
```

```python
#Анализируем результаты и выставляем баллы
import nbformat as nbf
import json

with open("generated/selection.txt", "r") as studentfile:
    studlistdict = json.load(studentfile)

groupstocount = ['4','6']
taskssolved = [0]*len(tasks)
#print(len(taskssolved))
for studkey in studlistdict:
    stname = studkey
    student = studlistdict[studkey]
    stgroup = student["group"]
    if stgroup in groupstocount:
        fullname = student['fio']
        #print(stname)
        nbfile = 'collected'+str(stgroup)+'/'+asmname+'_'+stname+'.ipynb'
        taskgroupidcs =  [problem["problemidx"] for problem in student["problems"] ] #human-adapted numbers of students tasks starting from 1
        taskvariantidcs =  [problem["variantidx"] for problem in student["problems"] ] #human-adapted numbers of students variant starting from 1
        taskparams = [problem["params"] for problem in student["problems"] ]
        tasks = [globaltasks[groupidx-1][variantidx-1] for groupidx, variantidx in zip(taskgroupidcs,taskvariantidcs)]
        
        cells = nbf.read(nbfile, nbf.NO_CONVERT)['cells']
        #print(cells)
        while len(cells)>0 and cells[-1]['source']=='':
            cells = cells[:-1]
        res = eval( cells[-1]['source'])
                       
        trueres = [tasks[taskidx]['answer'](**taskparams[taskidx]) for taskidx in range(len(tasks))] 
        
        reltolerance = 0.01
        mark = []
        for idx in res:            
            if trueres[idx-1] == 0:
                mark.append (abs(res[idx])<reltolerance)     
            else: 
                mark.append( abs((trueres[idx-1]-res[idx])/trueres[idx-1])<reltolerance )
            if mark[-1]==1:
                #print(taskidcs[idx-1]-1)
                taskssolved[taskgroupidcs[idx-1]-1] += 1
        #print (fullname, res)
        #print (taskidcs)
        #print (trueres)
        #print (mark)
        finalmarks = [str(idx+1)+"("+str(taskgroupidcs[idx])+':'+str(taskvariantidcs[idx]) +")"+('+' if mark[idx] else '-') for idx in range(len(taskgroupidcs))  ]
        total = sum(mark)
        if total!=0:
           print (fullname, " ".join(finalmarks), "итого:", total)
        #print (total)
        
print ("correct answers: ",taskssolved)
   
```
